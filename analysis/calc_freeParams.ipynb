```python
%autoreload 2

from cndlles.torch_util import *
from cndlles.torch_arch import *
import torch
```

```python
N = 4
size = 3
Ri_pct = 0.25
shape = (9, size, size)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
Nhid = [32,16,8]  # Multiplicity of hidden layers
model=CNDNN(Nhid, Ri_pct=Ri_pct, N=N, size=size).float().to(device)
```

```python
print('Total:')
print(sum(p.numel() for p in model.parameters() if p.requires_grad))
```

```python
model
```

```python
layer = model.u_in
print(layer)
for p in layer.parameters():
    if p.requires_grad:
        print(str(p.numel()) + " free parameters")
print()

layer = model.Ri_in
print(layer)
for p in layer.parameters():
    if p.requires_grad:
        print(str(p.numel()) + " free parameters")
print()

for layer in model.ops:
    print(layer)
    for p in layer.parameters():
        if p.requires_grad:
            print(str(p.numel()) + " free parameters")
    print()

layer = model.change_basis
print(layer)
for p in layer.parameters():
    print(p.shape)
    print(p.requires_grad)
```
